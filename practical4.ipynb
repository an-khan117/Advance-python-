```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 4, 2]

# Multiple colors for bars
colors = ['purple', 'blue', 'green', 'orange', 'red']

plt.title("Bar Plot")
plt.bar(x, y, color=colors, label="Bar Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(loc=2)
plt.show()
```
